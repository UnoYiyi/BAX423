### Import Packages

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Read Data

```python
df = pd.read_csv('./Small-Book Reviews from Amazon.csv', header = None)
text = df[1]
text
```

### BoW

```python
count = CountVectorizer()
bag = count.fit_transform(text)

#Fitting the CountVectorizer does sthe following: 1) tokenizing and 2) building the vocabulary. 
#Vocabulary can be accessed through the vocabulary_ attribute.
```

### Print Vocabulary

```python
## Now let us print the contents of the vocabulary to get a better understanding of the underlying concepts:
print(count.vocabulary_)
```

```python
print(len(count.vocabulary_)) ## check how many words are there in total
```

### Document-term Matrix

```python
print(bag.toarray())
```

## Comparing Unigram and Bigram
### Unigram Parameter and Number of Feature

```python
feature_names = count.get_feature_names()
print(feature_names)
```

```python
number_of_features = len(feature_names)
print(number_of_features)
```

### Bigram Parameter and Number of Feature

```python
count_bi = CountVectorizer(ngram_range=(2, 2))
bag_bi = count_bi.fit_transform(df[1])

feature_names_bi = count_bi.get_feature_names()
print(feature_names_bi)
```

```python
number_of_features_bi = len(feature_names_bi)
print(number_of_features_bi)
```

## TF-IDF

```python
vect = TfidfVectorizer()
tfidf = vect.fit_transform(text)
```
